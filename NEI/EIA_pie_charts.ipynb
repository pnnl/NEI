```python
import matplotlib.pyplot as plt
```

```python
# Data for main heating fuels
main_heating_fuels = {
    'Natural gas': 62.71,
    'Electricity': 42.57,
    'Fuel oil or kerosene': 4.93,
    'Propane': 5.21,
    'Wood': 2.25,
    'Does not use heating equipment': 5.79,
}

# Data for heating equipment types
heating_equipment_types = {
    'Central warm-air furnace': 74.42,
    'Heat pump': 16.13,
    'Steam or hot water system': 9.29,
    'Ductless heat pump (mini-split)': 1.06,
    'Built-in electric units': 7.65,
    'Portable electric heaters': 3.03,
    'Heating stove burning wood': 1.94,
    'Does not use heating equipment': 5.79,
}
```

```python
def create_pie_chart(data, title):
    labels = list(data.keys())
    sizes = list(data.values())
    total = sum(sizes)

    plt.figure(figsize=(6, 4), dpi=300)
    patches, texts, autotexts = plt.pie(sizes, labels=[''] * len(labels), autopct='', startangle=140)
    plt.legend(patches, [f'{label} ({sizes[i]/total:.1%})' for i, label in enumerate(labels)], title="", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.setp(autotexts, size=8)  # Set the font size of percentages in the legend
    plt.show()
```

```python
create_pie_chart(main_heating_fuels, 'Primary Heating Fuel Type')
```

```python
create_pie_chart(heating_equipment_types, 'Primary Heating Equipment Type')
```

```python
plt.figure(figsize=(6, 4), dpi=300)
create_pie_chart(main_heating_fuels, 'Main Heating Fuels', 0)
create_pie_chart(heating_equipment_types, 'Heating Equipment Types', 1)
plt.tight_layout(pad=2.0)  # Adjust the spacing between subplots
plt.show()
```
